```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import re
```

```python
kagglepod = pd.read_csv('poddf.csv')

kagglepod = kagglepod.drop('index',1)
kagglepod = kagglepod.drop(kagglepod[(kagglepod.Rating_Volume == 'Not Found')].index)
kagglepod = kagglepod.drop(kagglepod[(kagglepod.Rating == 'Not Found')].index)
kagglepod = kagglepod.drop_duplicates('Name', 'first')
kagglepod['Rating_Volume'] = kagglepod['Rating_Volume'].astype(float)
kagglepod['Rating'] = kagglepod['Rating'].astype(float)
kagglepod
```

```python
genre_keys = [['1311', 'News & Politics'], ['26', 'Podcasts'], ['1479', 'Social Sciences'], ['1315', 'Science & Medicine'], ['1324', 'Society & Culture'], ['1302', 'Personal Journals'], ['1469', 'Language Courses'], ['1304', 'Education'], ['1320', 'Places & Travel'], ['1416', 'Higher Education'], ['1465', 'Professional'], ['1316', 'Sports & Recreation'], ['1303', 'Comedy'], ['1305', 'Kids & Family'], ['1439', 'Christianity'], ['1314', 'Religion & Spirituality'], ['1444', 'Spirituality'], ['1309', 'TV & Film'], ['1462', 'History'], ['1310', 'Music'], ['1478', 'Medicine'], ['1321', 'Business'], ['1412', 'Investing'], ['1420', 'Self-Help'], ['1307', 'Health'], ['1481', 'Alternative Health'], ['1417', 'Fitness & Nutrition'], ['1467', 'Amateur'], ['1480', 'Software How-To'], ['1318', 'Technology'], ['1448', 'Tech News'], ['1456', 'Outdoor'], ['1477', 'Natural Sciences'], ['1301', 'Arts'], ['1454', 'Automotive'], ['1323', 'Games & Hobbies'], ['1438', 'Buddhism'], ['1443', 'Philosophy'], ['1401', 'Literature'], ['1402', 'Design'], ['1410', 'Careers'], ['1470', 'Training'], ['1413', 'Management & Marketing'], ['1306', 'Food'], ['1406', 'Visual Arts'], ['1446', 'Gadgets'], ['1468', 'Educational Technology'], ['1405', 'Performing Arts'], ['1460', 'Hobbies'], ['1471', 'Business News'], ['1404', 'Video Games'], ['1450', 'Podcasting'], ['1473', 'National'], ['1325', 'Government & Organizations'], ['1461', 'Other Games'], ['1466', 'College & High School'], ['1459', 'Fashion & Beauty'], ['1476', 'Non-Profit'], ['1415', 'K-12'], ['1455', 'Aviation'], ['1464', 'Other'], ['1421', 'Sexuality'], ['1472', 'Shopping'], ['1475', 'Local'], ['1441', 'Judaism'], ['1440', 'Islam'], ['1474', 'Regional'], ['1463', 'Hinduism']]
genres = dict()
for genre in genre_keys:
    genres[genre[0]] = genre[1]
```

```python
gitpod = pd.read_csv('df_popular_podcasts.csv')
gitpod = gitpod.rename(columns={'Genre IDs':'Genres', 'Episode Count':'Episode_Count', 'Episode Durations':'Episode_Durations', 'iTunes URL':'iTunes_URL', 'Feed URL':'Feed_URL', 'Podcast URL':'Podcast_URL'})
```

```python
for index in gitpod.index:
    duration = gitpod['Episode_Durations'][index]
    if duration == '0': 
        gitpod['Episode_Durations'][index] = [0]
    else: 
        dlist = re.findall("\d+\.\d+", duration)
        dlist_float = [float(element) for element in dlist]
        gitpod['Episode_Durations'][index] = dlist_float
    
    genreID = gitpod['Genres'][index]
    genreID_list = re.findall("[0-9]+", genreID)
    expanded_genres = [genres[g] for g in genreID_list]
    gitpod['Genres'][index] = expanded_genres
```

```python
all_podcasts = pd.merge(kagglepod, gitpod, on="Name", how="left")

for i in all_podcasts.index: 
    single_genre = all_podcasts['Genre'][i]
    many_genres = all_podcasts['Genres'][i]
    if type(many_genres) == float: 
        all_podcasts['Genres'][i] = single_genre
    elif type(many_genres) == list and single_genre not in many_genres: 
        all_podcasts['Genres'][i] = many_genres + [single_genre]

# all_podcasts.drop('Genre', 1)
```
